### Background
Dalam era digital saat ini, pemasaran melalui influencer menjadi salah satu strategi utama bagi UMKM untuk meningkatkan jangkauan pasar dan penjualan. Namun, banyak UMKM masih menghadapi kesulitan dalam menentukan influencer yang tepat untuk diajak bekerja sama.

Permasalahan ini muncul karena tidak adanya sistem berbasis data yang dapat membantu mengevaluasi kualitas dan relevansi influencer secara objektif. Akibatnya, keputusan yang diambil sering kali bersifat subjektif dan berisiko menyebabkan pemborosan anggaran serta rendahnya Return on Investment (ROI).
### Identified Problems
Berdasarkan analisis awal, terdapat beberapa permasalahan yang dihadapi oleh UMKM dalam menjalankan kampanye influencer marketing:
1. Sulit mengidentifikasi influencer dengan audiens yang sesuai dengan target pasar.
2. Tidak tersedia metrik yang jelas untuk mengukur performa influencer.
3. Proses pemilihan influencer masih dilakukan secara manual.
4. Alokasi anggaran kampanye tidak optimal.
5. Tidak adanya sistem prediksi hasil kampanye (ROI).
### Problem Clustering
Permasalahan yang telah diidentifikasi kemudian dikelompokkan menjadi beberapa kategori utama:
1. Influencer Quality Problem
   - Engagement tidak mencerminkan performa sebenarnya
2. Influencer Matching Problem
   - Ketidaksesuaian antara niche influencer dan brand UMKM
   - Target audiens tidak tepat
3. Budget Optimization Problem
   - Distribusi anggaran tidak efisien
   - Tidak ada simulasi alokasi biaya
### Selected Main Problem

Masalah pendukung:

### Problem Statement
UMKM kesulitan dan membuang banyak waktu serta anggaran untuk menemukan influencer yang memiliki audiens relevan dan metrik asli akibat tidak adanya sistem pencocokan berbasis data objektif.

## 2. Business Questions

Berdasarkan problem statement yang telah didefinisikan, berikut adalah
pertanyaan bisnis yang dapat diukur secara kuantitatif menggunakan data:

---

### BQ-1 — Influencer Matching
**"Influencer dari kategori dan tier mana yang paling cost-efficient
untuk direkomendasikan kepada UMKM dengan anggaran terbatas?"**

- Variabel ukur : `base_rate`, `tier`, `kategori`, `ratio_post`
- Metrik        : Cost-per-Value Score = base_rate / median_rate_per_tier
- Target        : Identifikasi top 20% influencer dengan skor efisiensi tertinggi
- Relevansi     : Menjawab masalah *Influencer Matching Problem* —
                  membantu UMKM memilih influencer relevan sesuai budget

---

### BQ-2 — Rate Fairness
**"Apakah harga (base_rate) yang ditetapkan influencer sudah proporsional
terhadap tier-nya, atau ada yang overpriced / underpriced?"**

- Variabel ukur : `base_rate`, `tier`, `log_base_rate`
- Metrik        : Z-score base_rate dalam setiap tier
                  → Z > 1.5 = overpriced, Z < -1.5 = underpriced
- Target        : Flagging influencer anomali harga per tier
- Relevansi     : Menjawab *Budget Optimization Problem* —
                  UMKM tidak membayar lebih dari harga wajar

---

### BQ-3 — Addon Value
**"Seberapa besar nilai tambah (addon) yang ditawarkan influencer
dibandingkan base rate-nya, dan apakah ini konsisten antar kategori?"**

- Variabel ukur : `ratio_owning`, `ratio_boost`, `ratio_link`, `kategori`
- Metrik        : Rata-rata rasio addon per kategori
- Target        : Identifikasi kategori dengan addon paling kompetitif
- Relevansi     : Membantu UMKM memutuskan paket terbaik selain base rate

---

### BQ-4 — Tier Distribution
**"Apakah distribusi influencer per kategori sudah merata,
atau ada kategori yang didominasi satu tier saja?"**

- Variabel ukur : `kategori`, `tier`
- Metrik        : Crosstab proporsi tier × kategori (dalam %)
- Target        : Deteksi kategori dengan pilihan tier terbatas
- Relevansi     : Membantu UMKM tahu apakah ada cukup pilihan
                  di kategori yang mereka butuhkan

---

### BQ-5 — Budget Simulation (Pendukung)
**"Dengan anggaran X rupiah, kombinasi influencer mana yang dapat
memaksimalkan jangkauan kampanye?"**

- Variabel ukur : `base_rate`, `tier`, `kategori`
- Metrik        : Total influencer terjangkau dalam budget,
                  dengan variasi komposisi Mega+Nano vs full Micro
- Target        : Simulasi 3 skenario alokasi budget
- Relevansi     : Langsung menjawab *Budget Optimization Problem*
                  yang menjadi masalah pendukung proyek ini

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
import seaborn as sns
from google.colab import files

pd.set_option('display.float_format', lambda x: f'{x:,.0f}')
plt.rcParams['figure.figsize'] = (12, 5)
plt.rcParams['font.family']    = 'sans-serif'

print("✅ Library siap!")
```

```python

print("📁 Upload: List_KOL_Indorse_Management.xlsx")
files.upload()
print("\n📁 Upload: _DEV__daftar_kategori_dan_niche.xlsx")
files.upload()

df_kol_raw  = pd.read_excel('List_KOL_Indorse_Management.xlsx', sheet_name='ALL KOL', header=2)
dev_sheets  = pd.read_excel('_DEV__daftar_kategori_dan_niche.xlsx', sheet_name=None)
df_existing = dev_sheets['Lembar1']

print(f"\n✅ KOL raw    : {df_kol_raw.shape}")
print(f"✅ Existing   : {df_existing.shape}")
```

```python

df_kol_clean = df_kol_raw[df_kol_raw['IG'].notna()].copy()
existing_igs = set(df_existing['username_ig'].str.lower().str.strip())
df_new       = df_kol_clean[~df_kol_clean['IG'].str.lower().str.strip().isin(existing_igs)].copy()

print(f"Total KOL valid   : {len(df_kol_clean)}")
print(f"Sudah ada (skip)  : {len(df_kol_clean) - len(df_new)}")
print(f"KOL baru          : {len(df_new)}")
```

```python

def get_base_rate(row):
    try:
        val = float(row['Feeds'])
        if not np.isnan(val) and val > 0:
            return val
    except: pass
    try:
        val = float(row['TikTok.1'])
        if not np.isnan(val) and val > 0:
            return int(val / 1.20)
    except: pass
    return np.nan

def resolve_rate(raw_value, base_rate, multiplier):
    try:
        val = float(raw_value)
        if not np.isnan(val) and val > 0:
            return int(val), 'Asli'
    except: pass
    try:
        base = float(base_rate)
        if not np.isnan(base):
            return int(base * multiplier), 'Rumus'
    except: pass
    return np.nan, 'N/A'

def clean_username(val):
    if pd.isna(val): return np.nan
    return str(val).lstrip('@').strip().lower().replace(' ', '_')

rows = []
for _, row in df_new.iterrows():
    br = get_base_rate(row)
    story_v, _  = resolve_rate(row['Story (incl Link)'], br, 0.25)
    post_v,  _  = resolve_rate(row['Reels'],             br, 0.70)
    pp_v,    _  = resolve_rate(row['Feeds Carousel'],    br, 0.15)
    own_v,   _  = resolve_rate(row['+Owning Content'],   br, 0.50)
    boost_v, _  = resolve_rate(row['+Boost Ads'],        br, 0.30)
    link_v,  _  = resolve_rate(row['+Tap Link'],         br, 0.40)

    rows.append({
        'nama_influencer': clean_username(row['IG']),
        'niche'          : row['Kategori'] if pd.notna(row['Kategori']) else np.nan,
        'username_tiktok': clean_username(row['TikTok']),
        'username_ig'    : clean_username(row['IG']),
        'base_rate'      : int(br) if pd.notna(br) else np.nan,
        'story_rate'     : story_v,
        'post_rate'      : post_v,
        'pp_rate'        : pp_v,
        'addon_owning'   : own_v,
        'addon_boost'    : boost_v,
        'addon_link'     : link_v,
    })

df = pd.DataFrame(rows)
print(f"✅ Mapping selesai — shape: {df.shape}")
print(df.head(3))
```

```python

def clean_niche(val):
    if pd.isna(val) or str(val).strip() == '': return np.nan
    primary = str(val).split('\n')[0].lstrip('- ').strip()
    return primary

df['niche_raw']   = df['niche']
df['niche']       = df['niche'].apply(clean_niche)
df['niche_count'] = df['niche_raw'].apply(
    lambda x: len(str(x).split('\n')) if pd.notna(x) else 0
)

df['username_tiktok'] = df['username_tiktok'].fillna(df['username_ig'])

for c in ['addon_owning', 'addon_boost', 'addon_link']:
    df[c] = df[c].replace(0, np.nan)


multipliers = {
    'story_rate': 0.25, 'post_rate': 0.70, 'pp_rate': 0.15,
    'addon_owning': 0.50, 'addon_boost': 0.30, 'addon_link': 0.40
}
for col, mult in multipliers.items():
    mask = df[col].isnull() & df['base_rate'].notna()
    df.loc[mask, col] = (df.loc[mask, 'base_rate'] * mult).astype(int)


df = df.dropna(subset=['base_rate']).reset_index(drop=True)


rate_cols = ['base_rate','story_rate','post_rate','pp_rate',
             'addon_owning','addon_boost','addon_link']

def winsorize(series, factor=3.0):
    q1, q3 = series.quantile(0.25), series.quantile(0.75)
    iqr = q3 - q1
    return series.clip(lower=q1 - factor*iqr, upper=q3 + factor*iqr)

for c in rate_cols:
    df[c] = winsorize(df[c]).astype(int)

df['No'] = range(1, len(df) + 1)
df = df[['No','nama_influencer','niche','username_tiktok','username_ig'] + rate_cols]

print("✅ Cleaning selesai!")
print(f"   Shape           : {df.shape}")
print(f"   Missing values  :\n{df.isnull().sum()}")
print(df.head(3))
```

```python

df['log_base_rate'] = np.log1p(df['base_rate']).round(4)
df['ratio_story']   = (df['story_rate']   / df['base_rate']).round(4)
df['ratio_post']    = (df['post_rate']    / df['base_rate']).round(4)
df['ratio_pp']      = (df['pp_rate']      / df['base_rate']).round(4)
df['ratio_owning']  = (df['addon_owning'] / df['base_rate']).round(4)
df['ratio_boost']   = (df['addon_boost']  / df['base_rate']).round(4)
df['ratio_link']    = (df['addon_link']   / df['base_rate']).round(4)

df['tier'] = pd.cut(
    df['base_rate'],
    bins=[0, 1_000_000, 5_000_000, 15_000_000, 50_000_000, float('inf')],
    labels=['Nano', 'Micro', 'Mid', 'Macro', 'Mega']
)

print("✅ Feature engineering selesai!")
print(df[['username_ig','base_rate','log_base_rate','tier','ratio_story','ratio_post']].head(5))
```

```python
niche_normalize = {
    'Beatuy, Lfestyle (Female)'                     : 'Beauty, Lifestyle (Female)',
    'Lifestlye (Male)'                               : 'Lifestyle (Male)',
    'Komedi, Lifestyle (Male)'                       : 'Comedy, Lifestyle (Male)',
    'Comedy, Lifestyle Male'                         : 'Comedy, Lifestyle (Male)',
    'Sport & Health'                                 : 'Health & Sport',
    'Health, Education'                              : 'Health & Sport, Education',
    'Place review, family'                           : 'Place Review, Family',
    'Family, Pasutri'                                : 'Pasutri, Family',
    'Lifestyle, Fashion,'                            : 'Lifestyle (Female), Fashion',

    'Beauty, lifestyle (female)'                     : 'Beauty, Lifestyle (Female)',
    'Fashion, Lifestyle (female)'                    : 'Fashion, Lifestyle (Female)',
    'Fashion, lifestyle (female)'                    : 'Fashion, Lifestyle (Female)',
    'Lifestyle (female)'                             : 'Lifestyle (Female)',
    'Lifestyle (female), Edukasi (tips & trick)'     : 'Lifestyle (Female), Education',
    'Lifestyle (female), Fashion'                    : 'Lifestyle (Female), Fashion',
    'Lifestyle (female), comedy'                     : 'Lifestyle (Female), Comedy',
    'Lifestyle Female'                               : 'Lifestyle (Female)',
    'Lifestyle Female, Beauty'                       : 'Lifestyle (Female), Beauty',
    'Lifestyle Female, Mom n Kids, Premium Looks'    : 'Lifestyle (Female), Mom & Kids, Premium Look',
    'Lifestyle Male, Health & Sport'                 : 'Lifestyle (Male), Health & Sport',
    'Premium Look, Lifestyle Female'                 : 'Lifestyle (Female), Premium Look',
    'Beauty & fashion'                               : 'Beauty, Fashion',

    'Actress'                                        : 'Actor/Actress',
    'Couple'                                         : 'Pasutri',
    'Homeliving, Family'                             : 'Homeliving, Family',
    'FnB, Lifestyle (Male), Entrepreneur'            : 'FnB, Lifestyle (Male), Entrepreneur',
}

df['niche_clean'] = df['niche'].replace(niche_normalize)

print("✅ Normalisasi niche selesai!")
print(f"   Niche unik sebelum : {df['niche'].nunique()}")
print(f"   Niche unik sesudah : {df['niche_clean'].nunique()}")



def assign_kategori(niche_val):
    if pd.isna(niche_val):
        return 'Uncategorized'

    n = niche_val.lower()

    if 'mom'      in n or 'kids'     in n:  return 'Parenting'
    if 'cooking'  in n:                     return 'Food'
    if 'fnb'      in n:                     return 'Food'
    if 'place review' in n:                 return 'Travel'
    if 'homeliving'   in n:                 return 'Homeliving'
    if 'gaming'   in n:                     return 'Entertainment'
    if 'comedy'   in n:                     return 'Entertainment'
    if 'actor'    in n or 'actress'  in n:  return 'Entertainment'
    if 'health'   in n or 'sport'    in n:  return 'Health'
    if 'beauty'   in n:                     return 'Beauty'
    if 'fashion'  in n:                     return 'Fashion'
    if 'entrepreneur' in n:                 return 'Education'
    if 'education' in n:                    return 'Education'
    if 'pasutri'  in n or 'family'   in n:  return 'Lifestyle'
    if 'premium'  in n:                     return 'Lifestyle'
    if 'lifestyle' in n:                    return 'Lifestyle'

    return 'Uncategorized'

df['kategori'] = df['niche_clean'].apply(assign_kategori)

print("\n✅ Mapping kategori selesai!")
print("\n📊 Distribusi Kategori:")
print(df['kategori'].value_counts().to_string())

uncategorized = df[df['kategori'] == 'Uncategorized'][['username_ig', 'niche', 'niche_clean']]
if len(uncategorized) > 0:
    print(f"\n⚠️  {len(uncategorized)} baris Uncategorized (perlu cek manual):")
    print(uncategorized.to_string())
else:
    print("\n✅ Tidak ada Uncategorized!")


cols_order = [
    'No', 'nama_influencer', 'kategori', 'niche_clean',
    'username_tiktok', 'username_ig',
    'base_rate', 'story_rate', 'post_rate', 'pp_rate',
    'addon_owning', 'addon_boost', 'addon_link',
    'log_base_rate', 'ratio_story', 'ratio_post', 'ratio_pp',
    'ratio_owning', 'ratio_boost', 'ratio_link', 'tier'
]

df = df[cols_order].rename(columns={'niche_clean': 'niche'})

print(f"\n✅ Kolom final: {df.columns.tolist()}")
print(df[['username_ig', 'kategori', 'niche', 'tier']].head(10).to_string())
```

```python

fig, axes = plt.subplots(1, 2, figsize=(14, 5))
kat_counts = df['kategori'].value_counts()
colors = plt.cm.Set2.colors[:len(kat_counts)]
axes[0].barh(kat_counts.index, kat_counts.values, color=colors)
axes[0].set_title('Jumlah KOL per Kategori', fontweight='bold')
axes[0].set_xlabel('Jumlah KOL')
for i, v in enumerate(kat_counts.values):
    axes[0].text(v + 0.2, i, str(v), va='center', fontweight='bold')

avg_rate = df.groupby('kategori')['base_rate'].median().sort_values(ascending=True)
axes[1].barh(avg_rate.index, avg_rate.values / 1e6, color=colors[:len(avg_rate)])
axes[1].set_title('Median Base Rate per Kategori (Juta Rp)', fontweight='bold')
axes[1].set_xlabel('Median Base Rate (Juta Rp)')

plt.tight_layout()
plt.show()

pivot = pd.crosstab(df['kategori'], df['tier'])
plt.figure(figsize=(10, 6))
sns.heatmap(pivot, annot=True, fmt='d', cmap='YlOrRd', linewidths=0.5)
plt.title('Distribusi KOL: Kategori × Tier', fontweight='bold')
plt.tight_layout()
plt.show()
```

```python


fig, axes = plt.subplots(1, 2, figsize=(14, 5))

tier_counts = df['tier'].value_counts().sort_index()
axes[0].bar(tier_counts.index, tier_counts.values, color=['#74b9ff','#0984e3','#6c5ce7','#fd79a8','#e17055'])
axes[0].set_title('Distribusi Tier KOL', fontsize=13, fontweight='bold')
axes[0].set_xlabel('Tier'); axes[0].set_ylabel('Jumlah KOL')
for i, v in enumerate(tier_counts.values):
    axes[0].text(i, v + 0.3, str(v), ha='center', fontweight='bold')

axes[1].hist(df['base_rate'] / 1e6, bins=20, color='#0984e3', edgecolor='white')
axes[1].set_title('Distribusi Base Rate (juta Rp)', fontsize=13, fontweight='bold')
axes[1].set_xlabel('Base Rate (Juta Rp)'); axes[1].set_ylabel('Jumlah KOL')
plt.tight_layout(); plt.show()

plt.figure(figsize=(12, 5))
tier_order = ['Nano','Micro','Mid','Macro','Mega']
sns.boxplot(data=df, x='tier', y='base_rate', order=tier_order,
            palette=['#74b9ff','#0984e3','#6c5ce7','#fd79a8','#e17055'])
plt.title('Sebaran Base Rate per Tier', fontsize=13, fontweight='bold')
plt.ylabel('Base Rate (Rp)'); plt.xlabel('Tier')
plt.gca().yaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f'Rp {x/1e6:.0f}jt'))
plt.tight_layout(); plt.show()

plt.figure(figsize=(12, 5))
top_niche = df['niche'].value_counts().head(10)
sns.barplot(y=top_niche.index, x=top_niche.values, palette='Blues_r')
plt.title('Top 10 Niche KOL', fontsize=13, fontweight='bold')
plt.xlabel('Jumlah KOL'); plt.ylabel('')
plt.tight_layout(); plt.show()

plt.figure(figsize=(10, 7))
corr_cols = ['base_rate','story_rate','post_rate','pp_rate','addon_owning','addon_boost','addon_link']
corr = df[corr_cols].corr()
mask = np.triu(np.ones_like(corr, dtype=bool))
sns.heatmap(corr, mask=mask, annot=True, fmt='.2f', cmap='Blues',
            linewidths=0.5, vmin=0, vmax=1)
plt.title('Korelasi antar Rate', fontsize=13, fontweight='bold')
plt.tight_layout(); plt.show()


print("\n📊 Statistik Deskriptif Rate (dalam Rp):")
print(df[corr_cols].describe().applymap(lambda x: f'Rp {x:,.0f}'))

print("\n📊 Rata-rata Base Rate per Tier:")
print(df.groupby('tier', observed=True)['base_rate']
        .agg(['mean','median','count'])
        .applymap(lambda x: f'Rp {x:,.0f}' if isinstance(x, float) else x))
```
